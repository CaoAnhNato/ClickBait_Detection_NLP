```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clickbait_data.csv')
df.head()
```

```python
# Biểu đồ phân phối class 0 (non-clickbait) và 1 (clickbait)
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='label', data=df, order=[0, 1], palette=['#4C78A8', '#F58518'])

# Đổi nhãn trục x cho dễ đọc
ax.set_xticklabels(['0 - non-clickbait', '1 - clickbait'])
ax.set_xlabel('Class')
ax.set_ylabel('Số lượng')
ax.set_title('Phân phối nhãn clickbait')

# Hiển thị số lượng trên từng cột
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom', fontsize=10
    )

plt.tight_layout()
plt.show()
```
